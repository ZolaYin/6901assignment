### 2) Re-read the data in such a way that all date columns are identified as dates and the earthquake ID is used as the index

Verify that this worked using the `head` and `info` functions.


```python
df = pd.read_csv('http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv', parse_dates=['time'])
df = df.set_index('id')
```

### 3) Use `describe` to get the basic statistics of all the columns

Note the highest and lowest magnitude of earthquakes in the databse.

```python
basic_stats = df.describe()
print(basic_stats)

highest_magnitude = basic_stats.loc['max', 'mag']
lowest_magnitude = basic_stats.loc['min', 'mag']

print("highest magnitude：", highest_magnitude)
print("lowest magnitude：", lowest_magnitude)
```

### 4) Use `nlargest` to get the top 20 earthquakes by magnitude

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.nlargest.html

```python
top_20_earthquakes = df.nlargest(20, 'mag')

top_20_earthquakes
```

Examine the structure of the `place` column. The state / country information seems to be in there. How would you get it out?

### 5) Extract the state or country using Pandas [text data functions](https://pandas.pydata.org/pandas-docs/stable/text.html)

Add it as a new column to the dataframe called `country`. Note that some of the "countries" are actually U.S. states.

```python
df['country'] = df.place.str.split(', ').str[-1]
df
```

### 6) Display each unique value from the new column

### 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger and

```python
filtered_df = df[df['mag'] >= 4]
print(filtered_df.head())
```

### 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes

Location name on the x axis, Earthquake count on the y axis

```python
earthquake_counts = filtered_df['country'].value_counts()

top_5_locations = earthquake_counts.head(5)

plt.figure(figsize=(10, 6))
top_5_locations.plot(kind='bar')

plt.title('Top 5 Locations with the Most Earthquakes (Magnitude > 4)')
plt.xlabel('Location')
plt.ylabel('Earthquake Count')

plt.show()
```

### 9) Make a histogram the distribution of the Earthquake magnitudes

https://pandas.pydata.org/pandas-docs/version/0.23/generated/pandas.DataFrame.hist.html
https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html

Do one subplot for the filtered and one for the unfiltered dataset.
Use a Logarithmic scale. What sort of relationship do you see?

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

df['mag'].hist(ax=ax[0], bins=20, log=True)
ax[0].set_title('All Earthquakes')
ax[0].set_xlabel('Magnitude')
ax[0].set_ylabel('Frequency (Log Scale)')

filtered_df['mag'].hist(ax=ax[1], bins=20, log=True)
ax[1].set_title('Earthquakes with Magnitude > 4')
ax[1].set_xlabel('Magnitude')
ax[1].set_ylabel('Frequency (Log Scale)')

plt.show()
```

First, it is apparent that the frequency of low-magnitude earthquakes is much higher than that of high-magnitude earthquakes. As the magnitude increases, the frequency of earthquakes decreases rapidly. This distribution presents a typical inverse relationship between magnitude and frequency, meaning that low-magnitude earthquakes occur more frequently, while high-magnitude earthquakes are relatively rare.

The chart titled "Earthquakes with Magnitude > 4" includes only earthquakes with a magnitude greater than 4. This histogram shows that, when considering earthquakes above magnitude 4, the frequency of earthquakes decreases as the magnitude increases. Although the frequency decreases, the rate of decrease appears to be slower than for the histogram including all magnitudes, and there is some fluctuation.

### 11) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

Use a two-column subplot with both the filtered and unfiltered datasets. Color the points by magnitude. Make it pretty

What difference do you note between the filtered and unfiltered datasets?

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6), sharex=True, sharey=True)

sc1 = axes[0].scatter(df['longitude'], df['latitude'], c=df['mag'], cmap='viridis')
axes[0].set_title('All Earthquakes')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')

cbar1 = fig.colorbar(sc1, ax=axes[0], orientation='vertical', fraction=0.046, pad=0.04)
cbar1.set_label('Magnitude')

sc2 = axes[1].scatter(filtered_df['longitude'], filtered_df['latitude'], c=filtered_df['mag'], cmap='viridis')
axes[1].set_title('Earthquakes (Magnitude > 4)')
axes[1].set_xlabel('Longitude')
axes[1].set_ylabel('Latitude')


cbar2 = fig.colorbar(sc2, ax=axes[1], orientation='vertical', fraction=0.046, pad=0.04)
cbar2.set_label('Magnitude')
```

First, the unfiltered dataset includes magnitudes ranging from 0 to 8, while the filtered dataset only displays earthquakes with magnitudes greater than 4. This results in a denser distribution of points in the unfiltered dataset as it includes earthquakes of all magnitudes; the points in the filtered dataset are more sparse, showcasing only the larger magnitudes. The filtered dataset stands out visually, making it easier to observe high-magnitude earthquakes. Overall, the filtered dataset facilitates a more focused analysis on the distribution and frequency of high-magnitude earthquakes.
